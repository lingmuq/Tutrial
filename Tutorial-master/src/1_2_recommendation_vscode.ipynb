# Visual Studio Code (VS Code) のおすすめプラグイン
---

## 目的
---
VS Codeのおすすめプラグインを示す．

1. Python <br>
PythonをVS Codeで使うために必要．

2. Python for VSCode <br>
VS CodeでPythonを効率的に利用可能らしい．

3. vscode-icons <br>
xplorerの各ディレクトリやファイルにアイコンがついて見やすい．

4. Bracket Pair Colorizer 2 <br>
かっこを色付きで表現できる．どのかっこ同士が対応しているかが分かりやすくておすすめ．

5. Path Autocomplete <br>
パスの入力候補を自動で提示してくれる．

6. Code Runner <br>
簡単なテストや実行確認を行うことができる．

7. Python preview <br>
変数や配列の中身を表示できる．
