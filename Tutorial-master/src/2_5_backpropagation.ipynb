## 変数定義

入力値: $x_1$,$x_2$,$x_3$

$X=[x_1,x_2,x_3]^{T}\in \mathbb{R}^{1 \times 3}$

```python
x = np.array([0.5, 0.1, -0.2])
```

目標値（求めたい値）:$t$

```python
t = 0.6
```

学習率: $\eta$

```python
learnrate = 0.5
```

1層目と2層目の間の重み: $\boldsymbol{w}^{(1)}$

$\boldsymbol{w}^{(1)}=[w_{11}^{(1)},w_{21}^{(1)};w_{12}^{(1)},w_{22}^{(1)};w_{13}^{(1)},w_{23}^{(1)}] \in \mathbb{R}^{3 \times 2}$

```python
w1 = np.array([[0.5, -0.6],[0.1, -0.2],[0.1, 0.7]])
```

2層目と3層目の間の重み: $\boldsymbol{w}^{(2)}$

$\boldsymbol{w}^{(2)}=[w_{11}^{(2)};w_{21}^{(2)}] \in \mathbb{R}^{2 \times 1}$

```python
w2 = np.array([0.1, -0.3])
```

バイアス：$\boldsymbol{b}^{(1)}$と$b_{1}^{(2)}$

```python
b1 = np.array([0,0])
b2 = np.array(0)
```

シグモイド関数: $\sigma(x) = \frac{1}{1-\exp(-x)}$

```python
#define sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

入力として$\boldsymbol{x_{1}}$,$\boldsymbol{x_{2}}$,$\boldsymbol{x_{3}}$を与えたときに，中間層での出力結果$\boldsymbol{a}^{(1)}$を求める．<br>

\begin{equation}
\boldsymbol{a}^{(1)}=f(\boldsymbol{z^{(1)}})
\hspace{10pt}\mathrm{where}\hspace{10pt}\boldsymbol{z^{(1)}}=X\boldsymbol{w}^{(1)}+\boldsymbol{b}^{(1)}
\tag{1}    
\end{equation}




式(1)を解くために，$\boldsymbol{z^{(1)}}=X\boldsymbol{w}^{(1)}+\boldsymbol{b}^{(1)}$を計算する．

```python
z1 = np.dot(x, w1)+b1
print(z1)
```

次に，$\boldsymbol{a}^{(1)}=\sigma(\boldsymbol{z^{(1)}})$を解く．

```python
a1 = sigmoid(z1)
print(a1)
```

次に，Step. 1で求めた$\boldsymbol{a}^{(1)}$を入力として，出力結果$a_{1}^{(2)}$を求める．

\begin{equation}
a_{1}^{(2)}=f(z_{1}^{(2)})
\hspace{10pt}\mathrm{where}\hspace{10pt}z_{1}^{(2)}=\boldsymbol{a}^{(1)}\boldsymbol{w}^{(2)}+b_{1}^{(2)}
\tag{2}    
\end{equation}

式(2)を解くために，$z_{1}^{(2)}=\boldsymbol{a}^{(1)}\boldsymbol{w}^{(2)}+b_{1}^{(2)}$を計算する．

```python
z2 = np.dot(a1, w2)+b2
print(z2)
```

次に，$a_{1}^{(2)}=f(z^{(2)})$を解く．

```python
a2 = sigmoid(z2)
print(a2)
```

最終的な出力$y$を求める．
\begin{equation}
y=a_{1}^{(2)}
\tag{3}
\end{equation}

```python
y=a2
```

## Step. 1 パラメータ$\boldsymbol{w}^{(2)}$と$b_{1}^{(2)}$の更新

<left>
<img src=https://onedrive.live.com/embed?resid=B8367B1E7E59C516%21578395&authkey=%21AIx5RMbeCKQ0t0s&width=3090&height=1412
width=90%>
<br>
</left>

<left>
<img src=https://onedrive.live.com/embed?resid=B8367B1E7E59C516%21578396&authkey=%21AI9rlyUJdKx993Y&width=3176&height=1790
width=90%>
<br>
</left>

$\boldsymbol{w}^{(2)}$についての$L$の勾配を求める．

\begin{equation}
\frac{\partial L}{\partial \boldsymbol{w}^{(2)}}=\frac{\partial L}{\partial a_{1}^{(2)}}\frac{\partial a_{1}^{(2)}}{\partial z_{1}^{(2)}} \frac{\partial z_{1}^{(2)}}{\partial \boldsymbol{w}^{(2)}}
\tag{5}    
\end{equation}

ここで3つの偏微分をそれぞれ計算する．<br>

\begin{equation}
\begin{split}
\frac{\partial L}{\partial a_{1}^{(2)}} &=\frac{\partial}{\partial a_{1}^{(2)}} (t - a_{1}^{(2)})^{2}\\
&= -2(t-a_{1}^{(2)})
\end{split}
\end{equation}
<br>


```python
dLda2 = -2 * (t-a2) # (t-a2) 
print(dLda2)
```


\begin{equation}
\begin{split}
\frac{\partial a_{1}^{(2)}}{\partial z_{1}^{(2)}} &=\frac{\partial}{\partial z_{1}^{(2)}} \sigma(z_{1}^{(2)})\\
&= \frac{\partial}{\partial z_{1}^{(2)}} \frac{1}{1+\exp(-z_{1}^{(2)})}\\
&= \frac{\partial}{\partial z_{1}^{(2)}} \frac{exp(-z_{1}^{(2)})}{(1+\exp(-z_{1}^{(2)}))^{2}}\\
&= \sigma(z_{1}^{(2)})(1-\sigma(z_{1}^{(2)}))\\
&=a_{1}^{(2)}(1-a_{1}^{(2)})
\end{split}
\end{equation}
<br>



```python
da2dz2 = a2 * (1-a2)
print(da2dz2)
```

\begin{equation}
\begin{split}
\frac{\partial z_{1}^{(2)}}{\partial \boldsymbol{w}^{(2)}} &= \frac{\partial}{\partial \boldsymbol{w}^{(2)}} (\boldsymbol{a}^{(1)} \boldsymbol{w}^{(2)} + b_{1}^{(2)})\\
&= \boldsymbol{a}^{(1)}
\end{split}
\end{equation}
<br>


```python
dz2dw2 = a1
print(dz2dw2)
```

つまり，
\begin{equation}
\begin{split}
\frac{\partial L}{\partial \boldsymbol{w}^{(2)}}&=\frac{\partial L}{\partial a_{1}^{(2)}}\frac{\partial a_{1}^{(2)}}{\partial z_{1}^{(2)}} \frac{\partial z_{1}^{(2)}}{\partial \boldsymbol{w}^{(2)}}\\
&= -2(t- a_{1}^{(2)}) a_{1}^{(2)}(1-a_{1}^{(2)})\boldsymbol{a}^{(1)}
\end{split}
\tag{6}    
\end{equation}

```python
dLdw2 = dLda2 * da2dz2* dz2dw2
print(dLdw2)
```

上記の値が$\frac{\partial L}{\partial \boldsymbol{w}^{(2)}}$となる．<br>
これを学習率$\eta$でスケール変更し，パラメータ$\boldsymbol{w^{(2)}}$を変更する．
\begin{equation}
\boldsymbol{w^{(2)}} ← \boldsymbol{w^{(2)}} - \eta \frac{\partial L}{\partial \boldsymbol{w}^{(2)}}
\tag{7}    
\end{equation}


```python
delta_w2 = learnrate * dLdw2
print(delta_w2)
```

$b_{1}^{(2)}$の更新も同様に次の計算で行われる．
\begin{equation}
\frac{\partial L}{\partial b_{1}^{(2)}}=\frac{\partial L}{\partial a_{1}^{(2)}}\frac{\partial a_{1}^{(2)}}{\partial z_{1}^{(2)}} \frac{\partial z_{1}^{(2)}}{\partial b_{1}^{(2)}}
\tag{8}    
\end{equation}

\begin{equation}
\begin{split}
\frac{\partial z_{1}^{(2)}}{\partial b_{1}^{(2)}} &= \frac{\partial}{\partial b_{1}^{(2)}} (\boldsymbol{a}^{(1)} \boldsymbol{w}^{(2)} + b_{1}^{(2)})\\
&= 1
\end{split}
\end{equation}
<br>

```python
dz2db2 = 1
```

```python
dLdb2 = dLda2 * da2dz2 * dz2db2
print(dLdb2)
```

上記の値が$\frac{\partial L}{\partial b_{1}^{(2)}}$となる．<br>
これを学習率$\eta$でスケール変更し，パラメータ$b_{1}^{(2)}$を変更する．
\begin{equation}
b_{1}^{(2)} ← b_{1}^{(2)} - \eta\frac{\partial L}{\partial b_{1}^{(2)}}
\tag{9}    
\end{equation}

```python
delta_b2 = learnrate * dLdb2
print(delta_b2)
```

## Step. 2 パラメータ$\boldsymbol{w}^{(1)}$と$\boldsymbol{b}^{(1)}$の更新

<left>
<img src=https://onedrive.live.com/embed?resid=B8367B1E7E59C516%21578398&authkey=%21APeqDrVttd6AyKc&width=3090&height=1412
width=90%>
<br>
</left>

$\boldsymbol{w}^{(1)}$についての$L$の勾配を求める．

\begin{equation}
\frac{\partial L}{\partial \boldsymbol{w}^{(1)}}=\frac{\partial L}{\partial a_{1}^{(2)}}\frac{\partial a_{1}^{(2)}}{\partial z_{1}^{(2)}} \frac{\partial z_{1}^{(2)}}{\partial \boldsymbol{a}^{(1)}}\frac{\partial \boldsymbol{a}^{(1)}}{\partial \boldsymbol{z}^{(1)}}\frac{\partial \boldsymbol{z}^{(1)}}{\partial \boldsymbol{w}^{(1)}}\\
\tag{10}    
\end{equation}

先ほどの，$\boldsymbol{w^{(2)}}$の更新時に$\frac{\partial L}{\partial a_{1}^{(2)}}$と$\frac{\partial a_{1}^{(2)}}{\partial z_{1}^{(2)}}$は求めている．
\begin{equation}
\frac{\partial L}{\partial a_{1}^{(2)}} = -2(t-a_{1}^{(2)})
\end{equation}

\begin{equation}
\frac{\partial a_{1}^{(2)}}{\partial z_{1}^{(2)}} =a_{1}^{(2)}(1-a_{1}^{(2)})
\end{equation}
<br>


そのため，3つの偏微分をそれぞれ計算する．<br>
\begin{equation}
\begin{split}
\frac{\partial z_{1}^{(2)}}{\partial \boldsymbol{a}^{(1)}} &=\frac{\partial}{\partial \boldsymbol{a}^{(1)}} (\boldsymbol{a}^{(1)} \boldsymbol{w}^{(2)} + b_{1}^{(2)})\\
&= \boldsymbol{w}^{(2)}
\end{split}
\end{equation}
<br>

```python
dz2da1 = w2
print(dz2da1)
```





\begin{equation}
\begin{split}
\frac{\partial \boldsymbol{a}^{(1)}}{\partial \boldsymbol{z}^{(1)}} &= \frac{\partial}{\partial \boldsymbol{z}^{(1)}} \sigma(\boldsymbol{z^{(1)}})\\
&= \frac{\partial}{\partial \boldsymbol{z}^{(1)}} \frac{1}{1+\exp(-\boldsymbol{z}^{(1)})}\\
&=\boldsymbol{a}^{(1)}(1-\boldsymbol{a}^{(1)})
\end{split}
\end{equation}
<br>



```python
da1dz1 = a1*(1-a1)
print(da1dz1)
```

\begin{equation}
\begin{split}
\frac{\partial \boldsymbol{z}^{(1)}}{\partial \boldsymbol{w}^{(1)}} &= \frac{\partial}{\partial \boldsymbol{w}^{(1)}} (X\boldsymbol{w}^{(1)} + \boldsymbol{b}^{(1)})\\
&=X
\end{split}
\end{equation}
<br>



```python
dz1dw1 = x[:,None]
```

つまり，
\begin{equation}
\begin{split}
\frac{\partial L}{\partial \boldsymbol{w}^{(1)}}&=\frac{\partial L}{\partial a_{1}^{(2)}}\frac{\partial a_{1}^{(2)}}{\partial z_{1}^{(2)}} \frac{\partial z_{1}^{(2)}}{\partial \boldsymbol{a}^{(1)}}\frac{\partial \boldsymbol{a}^{(1)}}{\partial \boldsymbol{z}^{(1)}}\frac{\partial \boldsymbol{z}^{(1)}}{\partial \boldsymbol{w}^{(1)}}\\
&= -2(t-a_{1}^{(2)}) a_{1}^{(2)}(1-a_{1}^{(2)}) \boldsymbol{w}^{(2)} \boldsymbol{a}^{(1)}(1-\boldsymbol{a}^{(1)}) X
\end{split}
\tag{11}    
\end{equation}

```python
dLdw1 = dLda2 * da2dz2 * dz2da1 *da1dz1 * dz1dw1
print(dLdw1)
```

上記の値が$\frac{\partial L}{\partial \boldsymbol{w}^{(1)}}$となる．<br>
これを学習率$\eta$でスケール変更し，パラメータ$\boldsymbol{w^{(1)}}$を変更する．
\begin{equation}
\boldsymbol{w^{(1)}} ← \boldsymbol{w^{(1)}} - \Delta{ \boldsymbol{w}^{(1)}}
\tag{12}    
\end{equation}
ここで，$\Delta{ \boldsymbol{w}^{(1)}} = \eta \frac{\partial L}{\partial \boldsymbol{w}^{(1)}}$である．

```python
delta_w1 = learnrate * dLdw1
print(delta_w1)
```

$\boldsymbol{b}^{(1)}$の更新も同様に次の計算で行われる．
\begin{equation}
\frac{\partial L}{\partial \boldsymbol{b}^{(1)}}=\frac{\partial L}{\partial a_{1}^{(2)}}\frac{\partial a_{1}^{(2)}}{\partial z_{1}^{(2)}} \frac{\partial z_{1}^{(2)}}{\partial \boldsymbol{a}^{(1)}}\frac{\partial \boldsymbol{a}^{(1)}}{\partial \boldsymbol{z}^{(1)}}\frac{\partial \boldsymbol{z}^{(1)}}{\partial \boldsymbol{b}^{(1)}}
\tag{13}    
\end{equation}

\begin{equation}
\begin{split}
\frac{\partial \boldsymbol{z}^{(1)}}{\partial \boldsymbol{b}^{(1)}} &= \frac{\partial}{\partial \boldsymbol{b}^{(1)}} (X\boldsymbol{w}^{(1)} + \boldsymbol{b}^{(1)})\\
&= 1
\end{split}
\end{equation}
<br>

```python
dz1db1 = 1
```

```python
dLdb1 = dLda2 * da2dz2 * dz2da1 *da1dz1 * dz1db1
print(dLdb1)
```

上記の値が$\frac{\partial L}{\partial \boldsymbol{b}^{(1)}}$となる．<br>
これを学習率$\eta$でスケール変更し，パラメータ$\boldsymbol{b}^{(1)}$を変更する．
\begin{equation}
\boldsymbol{b}^{(1)} ← \boldsymbol{b}^{(1)} - \eta\frac{\partial L}{\partial \boldsymbol{b}^{(1)}}
\tag{14}    
\end{equation}

```python
delta_b1 = learnrate * dLdb1
print(delta_b1)
```

## 誤差逆伝播法による計算の効率化
すべてのパラメータの更新に$\frac{\partial L}{\partial a_{1}^{(2)}}\frac{\partial a_{1}^{(2)}}{\partial z_{1}^{(2)}}$は共有されるため，パラメータごとに最急降下法をそれぞれのパラメータに用いた更新よりも大幅に計算コストを削減できる．
