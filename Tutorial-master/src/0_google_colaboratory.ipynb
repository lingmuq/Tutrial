# Google Colaboratoryの使い方

## 目的
Google Colaboratoryの使い方を学ぶ．

## Google Colaboratory (Colab) とは
Colabは，Googleが提供するブラウザ上で Python を記述，実行できるJupyter Notebook環境です．
- 環境構築が不要
- GPU環境を使用可能
- 無料で利用可能

## Jupyter Notebook
Jupyter Notebookでは，コードや説明文を記述することが可能な「cell」を使用する．
「cell」には，次の2種類が存在する．
- Code cell: Pythonのコードを記述し，実行することが可能．
- Markdown cell: **Markdown記法**を用いて，説明文等を記載することが可能．
