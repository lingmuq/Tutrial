# Anacondaのcondaコマンド
---

Anacondaでよく使うコマンドを示す．

## Anacondaパッケージ用コマンド
```
conda update -n base conda
```

## 仮想環境用コマンド
- 空の仮想環境を構築 <br>
```
conda create -n [environment_name]
```
- 仮想環境を構築 <br>
```
conda create -n [environment_name] python=[version] anaconda
```

- 仮想環境を削除 <br>
```
conda remove -n [environment_name] --all
```

- 仮想環境の名前を変更（新しい名前でクローンを作成して古い環境を消す） <br>
```
conda create -n [new_environment_name] --clone [old_environment_name]
conda remove -n [old_environment_name] -all
```

- 指定した仮想環境に切り替え <br>
```
conda activate [environment_name]
```

- 指定した仮想環境を停止 <br>
```
conda deactivate [environment_name]
```

## パッケージ関連コマンド
- パッケージのインストール（バージョン指定なし） <br>
```
conda install [package_name]
```
- パッケージのインストール（バージョン指定あり） <br>
```
conda install [package_name]==*.*.*
```
- パッケージのアンインストール <br>
```
conda uninstall [package_name]
```
- パッケージを更新 <br>
```
conda update [package_name]
```
- パッケージをすべて更新 <br>
```
conda update --all
```
- パッケージを検索 <br>
```
conda search [package_name]
```
- インストール済みのパッケージをすべて表示 <br>
```
conda list
```
- 指定したインストール済みのパッケージを表示 <br>
```
conda list [package_name]
```
