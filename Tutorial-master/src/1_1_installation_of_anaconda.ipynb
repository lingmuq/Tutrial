次の手順に従いAnacondaをインストールする．<br>

## Python環境の構築
Anacondaは複数のPython環境を作成することができるため，プロジェクトごとに独立した環境でPythonコードを実行できる．<br>
これにより，異なるPythonのバージョンを用いるプログラムであっても，実行のたびにPythonをインストールし直すなどの必要がなく，実行環境を切り替えるだけでプログラムを実行することができる．<br>
AnacondaとVS Codeを用いた仮想環境の構築方法について示す．<br>

1. 「スタート」からフォルダ「Anaconda3 (64bit)」-> 「Anaconda Prompt」をクリックして実行する．
<br>
<br>
<left>
<img src="https://dsm01pap004files.storage.live.com/y4m9_4ekRF4_LBsvgGgU7FL95wqKG2ZcHfKfw0iMAIWXaeXgJRDnjJI7BXd_4IKnoa95nB64CRSmyc_HnWTISg_WPS-Xd-bNBDEFHDTbEu-r0CcCL7Yb4kiDh6yjOMifkWuHxlobWLuzN1rZyEu0-yzWeenTWKsGL6O3eRE8ePFr77-SA3NEhLMiTUdnq7W8a4l?width=634&height=480&cropmode=none" width=20%>
</left>
<br>

2. "base"の環境にいることをコマンド"conda info -e"を入力して確認する．<br>

```
(base) C:\Users\username>
(base) C:\Users\username> conda info -e
# conda enviroments:
#
base            * C:\Users\username\Anaconda3

```
*マークがついている，または，コマンドプロンプトの最初の（）の中に書いてある場所が現在の環境である．<br>
<span style="color: red; ">環境"base"には，パッケージや異なるバージョンのPythonをインストールしないよう</span>にしたいため，新しい環境を作成する．

3. 新しい環境をコマンド"conda create -n 環境名"を入力して作成する．<br>
ここでは，新たな環境"env_test"を作成する．
```
(base) C:\Users\username>
(base) C:\Users\username> conda create -n env_test
(base) C:\Users\username> conda info -e
# conda enviroments:
#
base            * C:\Users\username\Anaconda3
env_test          C:\Users\username\Anaconda3\envs\env_test
```

4. 新しい環境をコマンド"conda activate 環境名"を入力して現在の環境を切り替える．<br>
```
(base) C:\Users\username>
(base) C:\Users\username> conda activate env_test
(env_test) C:\Users\username> conda info -e
# conda enviroments:
#
base              C:\Users\username\Anaconda3
env_test        * C:\Users\username\Anaconda3\envs\env_test
```

5. 新たな環境"env_test"は空であるため，Pythonをコマンド"conda install パッケージ名"でインストールする．<br>
ここでは，コンピュータヴィジョンのためのライブラリopencvと計算用ライブラリnumpyをインストールすることで，opencvに依存するPythonのバージョンをインストールすることが可能である．
```
(base) C:\Users\username>
(env_test) C:\Users\username> conda install -c conda-forge opencv
(env_test) C:\Users\username> conda install numpy
(env_test) C:\Users\username> conda list
# packages in environment at C:\Users\katsu\anaconda3\envs\env_pytorch:
#
# Name                    Version                   Build  Channel
python                    3.9.7                h6244533_1
.
.
.
```
＊ "-c conda-forge"はパッケージの検索先チャンネルを指定している．普段は，入力しなくても問題ない．

11. pythonファイル"main.py"に，次のサンプルコードを入力して実行する．<br>

````python
# 必要なパッケージの読み込み
import numpy as np
import cv2

# 設定
scale = 1.0
color = (255, 255, 255)
thickness = 1

# 画像生成
img = np.zeros((256, 256, 3), np.uint8)

# 文字生成
text = "Hello"
cv2.putText(img, text, (100, 100), cv2.FONT_HERSHEY_SIMPLEX, scale, color, thickness, cv2.LINE_AA)

# 表示
cv2.imshow("Window", img)
print("任意のキーを入力してください．")
cv2.waitKey(0)
cv2.destroyAllWindows()
````
